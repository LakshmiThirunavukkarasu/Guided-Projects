```python
cv = CountVectorizer(max_features = 5000)
word_count_vector = cv.fit_transform(df['refined_sentence'])
```

```python
print(len(cv.get_feature_names()))
```

```python
print(cv.get_feature_names())
```

```python
tfidf_transformer = TfidfTransformer()
tf_idf_vector = tfidf_transformer.fit_transform(word_count_vector)
```

```python
tf_idf_vector_norm = normalize(tf_idf_vector,norm = 'l1',axis=1)
```

```python
#Number of topcis
no_of_topics = 5
model = NMF(n_components = no_of_topics, init='nndsvd' )
model.fit(tf_idf_vector_norm)
```

```python
def get_nmf_topics(model, n_top_words):
    feat_names = cv.get_feature_names()
    
    word_dict ={}
    for i in range(no_of_topics):
        words_ids= model.components_[i].argsort()[:-n_top_words-1:-1]
        words = [feat_names[key] for key in words_ids]
        word_dict['Topic #' + '{:02d}'.format(i+1)] = words
    return pd.DataFrame(word_dict)
```

```python
get_nmf_topics(model,10 )
```
