```python
import pandas as pd
import numpy as np
```

```python
df  = pd.read_csv("store_data.csv",header = None)
df.head(5)
```

```python
df = df.fillna(0)
ItemList = set(np.concatenate(df.values))
marketData = pd.DataFrame(columns =ItemList)
marketData.drop(0, axis = 1, inplace = True)
```

```python
list(marketData.columns)
```

```python
for item in ItemList:
    col_val = []
    x = str(item).strip()
    for i in range(len(df.values)):
        if (x in df.values[i]):
            col_val.append(1)
        else:
            col_val.append(0)
    marketData[item] = col_val
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#apriori?
```

```python
freq_itemsets = apriori(marketData, min_support=0.01,use_colnames=True)
```

```python
marketData.drop(0,inplace= True)
```

```python
freq_itemsets.sort_values(by = 'support', ascending= False)
```

```python
#assoiciation_rules?
```

```python
#rules = association_rules(freq_itemsets,metric = 'lift',min_threshold = 0.2)
rules = association_rules(freq_itemsets,metric = 'support',min_threshold = 0.2)
```
